```python
import requests
from bs4 import BeautifulSoup
import html5lib as html
import pandas as pd
#request
title_list = list()
posted_dates = list()
closing_dates = list()
base_url = 'https://hec.ac.mw/scholarships/'
#Lists above to store results as they come 
for page in range(1, 8):
    #pagination control on
    response = requests.get(base_url+str(page))
    soup = BeautifulSoup(response.text, 'html5lib')
    posted_el = soup.find_all('p', class_='scholar-meta')
    closing_els = soup.find_all('div', class_='scholar-content-box')
    title_elements = soup.find_all('h4', class_='scholarship-title')
    for item in title_elements:
        el = item.find('a').text
        title_list.append(el)
    for item in posted_el:
        key =  item.text.strip()#remove trailing spaces
        posted_dates.append(key.split()[2]+'-'+key.split()[3]+'-'+key.split()[4])
    for item in closing_els:
        key = item.find('p', {'class':None, 'style':None})
        key = key.text.strip().split()
        index = key.index('Closes')
        closing_dates.append(key[index+1]+'-'+key[index+2]+'-'+key[index+3])
df = pd.DataFrame({'Scholarship':title_list, 'Posting Date':posted_dates, 'Closing Date':closing_dates})
#clean dates
df['Posting Date'] = pd.to_datetime(df['Posting Date'])
df['Closing Date'] = pd.to_datetime(df['Closing Date'])

df.to_csv('hec-scholarship.csv')
df
#tree

```
